### Data cleaning

```python
# drop unwanted column
book_df.drop(columns=['index'],inplace=True)
```

```python
# check for the number of the missing values
book_df.isna().sum()
```

```python
book_df.columns
```

```python
# lets deal with the missing value in publishing year
book_df['Publishing Year'].unique()
```

```python
# filter the record where the publishing year is empty
book_df[book_df['Publishing Year'].isna()]
```

***Since the number of records with missing value is less than 5%, i can drop the record.***

```python
#dropping the record with the missing values in the year column
book_df.dropna(subset=['Publishing Year'], inplace=True)
```

```python
book_df.info()
```

```python
# lets deal with inconsistency in the year.A year should not be negative.
# books with unclean publishing year
book_df[book_df['Publishing Year']< 0]
```

```python
# This are books that, the publishing year are greater than 0, i.e greater than minus.
book_df[book_df['Publishing Year']> 0]
```

```python
#books with valid publishing year.Saved into a new var after dropped those with unclean publishin year
book_df = book_df[book_df['Publishing Year']> 0]
```

```python
book_df
```

```python
book_df.info()
```

```python
book_df[book_df['Publishing Year']< 0]
```

***Book Name: lets clean Book Name column up now***

```python
book_df['Book Name'].isna().sum()
```

```python
# fill the missing books with untitled book
book_df['Book Name'].fillna('untitled book', inplace=True)
```

```python
book_df['Book Name'].isna().sum()
```

***Language Code: lets clean Language Code column up now***

```python
book_df['language_code'].isna().sum()
```

```python
book_df['language_code'].unique()
```

```python
# fill the missing language code with mode
mode_lang_code = book_df['language_code'].mode()[0]
book_df['language_code'].fillna(mode_lang_code, inplace=True)
```

```python
book_df['language_code'].isna().sum()
```

```python
# lets check if all datasets are clean now
book_df.info()
```

```python
# lets check for what Author Rating looks like - confirm the datatype
book_df['Author_Rating'].unique()
```

```python
# There is inconsistency here; genre-fiction
book_df['genre'].unique()
```

```python
# lets rename genre fiction with fiction
book_df['genre'] = book_df['genre'].replace('genre fiction','fiction')
```

```python
# our result
book_df['genre'].unique()
```

```python
# Just to check if that column is clean
book_df['gross sales'].min()
```

1. How many Unique BOOKS do we have?

```python
num_of_books = book_df['Book Name'].nunique()
print(f"The number of unique books available is:{num_of_books} pieces")
```

#### 2. Number of Authors

```python
num_of_authors = book_df['Author'].nunique()
print(f"The number of Authors available is:{num_of_authors}")
```

#### 3. Number of Publishers

```python
num_of_publishers = book_df['Publisher '].nunique()
print(f"The number of Publishers available is:{num_of_publishers}")
```

#### 4. Total Gross Sales

```python
total_of_gross_sales = book_df['gross sales'].sum()
print(f"The number of gross sales available is:${total_of_gross_sales}")
```

#### 5. Total Unit sold

```python
total_unit_sold = book_df['units sold'].sum()
print(f"The number of Total Unit Sold is:{total_unit_sold}")
```

#### 6. Total Publisher Revenue

```python
total_publisher_revenue = book_df['publisher revenue'].sum()
print(f"The Total Publisher Revenue is:${total_publisher_revenue}")
```

#### 7. Average Sales Price

```python
avg_sales_price = book_df['sale price'].mean()
print(f"The Average sales Price is:${avg_sales_price }")
```
